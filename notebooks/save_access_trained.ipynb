```python
import pandas as pd
import psycopg2
from psycopg2.extras import execute_values
```

```python
PATH = './parquets'
```

```python
DB_PARAMS = {
    "dbname": "db",
    "user": "user",
    "password": "pass",
    "host": "localhost",
    "port": 5416
}
```

```python
INSERT_SQL = """INSERT INTO "access" (id_user, id_news, data_acesso, engagement) VALUES %s ON CONFLICT (id) DO NOTHING"""
```

```python
CHUNK_SIZE = 1000
```

```python
access = pd.read_parquet(f'{PATH}/df_access.parquet')
```

```python
access['data_acesso'] = pd.to_datetime(access['timestampHistory'], unit='ms').dt.tz_localize(None)
```

```python
access.rename(columns={'userId': 'id_user', 'newsId': 'id_news', 'engagement_score' : 'engagement'}, inplace=True)
```

```python
access = access[['id_user', 'id_news', 'data_acesso', 'engagement']]
```

```python
access.head()
```

```python
try:
    conn = psycopg2.connect(**DB_PARAMS)
    cur = conn.cursor()

    for i in range(0, len(access), CHUNK_SIZE):
        chunk = access.iloc[i:i + CHUNK_SIZE]
        values = [(x[0],x[1],x[2],x[3]) for x in chunk.to_numpy()]
        execute_values(cur, INSERT_SQL, values)
        conn.commit()
        print(f"Chunk {i // CHUNK_SIZE + 1} inserido com sucesso.")

    cur.close()
    conn.close()
    print("Insero concluda!")

except Exception as e:
    print(f"Erro: {e}")
```
